# Problem 1
Find the total number of crimes recorded in the CRIME table.

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
SELECT COUNT("primary_type") AS total_crime FROM "CrimeData";     
```

# Problem 2
List community area names and numbers with per capita income less than 11000.

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB

SELECT "community_area_name", "per_capita_income"
FROM "CensusData"
WHERE "per_capita_income" < 11000
GROUP BY "community_area_name", "per_capita_income"
ORDER BY "per_capita_income";             
```

# Problem 3
List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis)

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
SELECT "case_number", "description" FROM "CrimeData" 
WHERE "description" ILIKE '%MINOR%'
GROUP BY "case_number", "description"
ORDER BY "description";        
```

# Problem 4
List all kidnapping crimes involving a child?

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
SELECT "case_number", "primary_type" FROM "CrimeData"
WHERE "primary_type" ILIKE '%KIDNAPPING%' 
GROUP BY "case_number", "primary_type"
ORDER BY "primary_type";        
```

# Problem 5
List the kind of crimes that were recorded at schools. 

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
SELECT  "primary_type","location_description"FROM "CrimeData" 
WHERE "location_description" ILIKE '%SCHOOL%'
GROUP BY "primary_type","location_description"
ORDER BY "location_description";        
```

# Problem 6
List the type of schools along with the average safety score for each type.
 

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
 SELECT "elementary, middle, or high school" AS "types_of_schools",
       AVG("safety_score") AS "average_safety_score"
       FROM  
       "SchoolsData"
       GROUP BY 
       "elementary, middle, or high school"
       ORDER BY 
       "average_safety_score" DESC;         
```

# Problem 7
List 5 community areas with highest % of households below poverty line.

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
SELECT  "community_area_name", MAX("percent_households_below_poverty") as highest_percent_households_below_poverty
FROM  "CensusData"
GROUP BY  "community_area_name"
ORDER BY highest_percent_households_below_poverty DESC
LIMIT 5;             
```

# Problem 8
Which community area is most crime prone? Display the coumminty area number only.

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
SELECT
       "community_area_number" AS most_crime_prone,
COUNT(*) AS crime_count
       FROM
       "CrimeData"
GROUP BY
       "community_area_number"
ORDER BY
       crime_count DESC
LIMIT 1;         
```

# Problem 9
Use a sub-query to find the name of the community area with highest hardship index.

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
SELECT "community_area_name","hardship_index" AS highest_hardship_index
FROM "CensusData"
WHERE"hardship_index" = (select MAX("hardship_index")from"CensusData");         
```

# Problem 10
Use a sub-query to %%sql postgresql://postgres:hdetermine the Community Area Name with most number of crimes?


```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
SELECT SD."community_area_number", SD."community_area_name"
FROM "SchoolsData" SD
JOIN (
    SELECT "community_area_number", COUNT(*) AS total_crimes
    FROM "CrimeData"
    GROUP BY "community_area_number"
    HAVING "community_area_number" = (
        SELECT MAX("community_area_number") FROM "CrimeData"
    )
) AS max_crime_area ON SD."community_area_number"::TEXT = max_crime_area."community_area_number";
```
