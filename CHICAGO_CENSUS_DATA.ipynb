# Problem 2
How many community areas in Chicago have a hardship index greater than 50.0?

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
SELECT "community_area_name"
FROM "CensusData"
WHERE hardship_index>50;
```

# Problem 4

# Problem 5
Which Chicago community areas have per-capita incomes greater than $60,000?

```python
%%sql postgresql://postgres:helenmario@localhost/ibmsqlassDB
SELECT community_area_name
FROM "CensusData"
WHERE per_capita_income>60000;        
```

# Problem 6
Create a scatter plot using the variables per_capita_income_ and hardship_index. Explain the correlation between the two variables.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import psycopg2

# Database connection details
db_connection_str = 'postgresql://postgres:helenmario@localhost/ibmsqlassDB'
# Establishing the connection
conn = psycopg2.connect(db_connection_str)

# SQL query to retrieve data
sql_query = 'SELECT per_capita_income, hardship_index FROM "CensusData"'

# Fetching the data into a DataFrame
income_vs_hardship_df = pd.read_sql_query(sql_query, conn)

# Closing the connection
conn.close()

# Plotting
plot = sns.jointplot(x='per_capita_income', y='hardship_index', data=income_vs_hardship_df)
plt.show()
```
